```python
import pandas as pand

users = pand.read_table(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|', index_col='user_id')

occupationofusers = users.set_index(["occupation"])
      
meanage = occupationofusers["age"].groupby('occupation').mean()
meanage
```

```python
gendertotal = users['gender'].value_counts()
genderoccupation = users.groupby('gender')['occupation'].value_counts()
sex_ratio = genderoccupation['M']/genderoccupation['F']
print(sex_ratio.sort_values(ascending=False))
```

```python
users.groupby('occupation').age.agg('max', 'min')
```

```python
users.groupby(['occupation','gender'])['age'].mean()
```

```python
gendertotal = users['gender'].value_counts()
occupationofgender = users.groupby('gender')['occupation'].value_counts()
male_sex_ratio = occupationofgender['M']/(occupationofgender['M'] + occupationofgender['F'])*100
female_sex_ratio = occupationofgender['F']/(occupationofgender['M'] + occupationofgender['F'])*100
print(male_sex_ratio, female_sex_ratio)
```

Housing
Step 1. Import the necessary libraries

Step 2. Create 3 differents Series, each of length 100, as follows:

• The first a random number from 1 to 4

• The second a random number from 1 to 3

• The third a random number from 10,000 to 30,000

Step 3. Create a DataFrame by joinning the Series by column

Step 4. Change the name of the columns to bedrs, bathrs, price_sqr_meter

Step 6. Ops it seems it is going only until index 99. Is it true?

```python
import pandas as pand
import numpy as nump
S1=pand.Series(nump.random.randint(1,4,100))
S2=pand.Series(nump.random.randint(1,3,100))
S3=pand.Series(nump.random.randint(10000, 30000,100))
print(S1)
print(S2)
print(S3)
```

```python
d1=S1.to_frame(name='series1')
d2=S2.to_frame(name='series2')
d3=S3.to_frame(name='series3')

finaldata =pand.concat([d1, d2, d3], 
                  axis = 1)
display(finaldata)
```

```python
finaldata.rename(columns={'series1':'bedrs','series2':'bathrs', 'series3':'price_sqr_meter'}, inplace=True)
display(finaldata)
```

```python
bigcolumn = pand.concat([S1,S2,S3], axis=0)
d1column=bigcolumn.to_frame(name='bigcolumn')
display(d1column)
print(type(d1column))
```

```python
d1column.reset_index(drop=True, inplace= True)
display(d1column)
```
